```python
import pandas as pd
from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
data = pd.read_csv(file_name)
print(data.head())

# 종속 변수와 독립 변수 설정
y = data['chimsoo']
X_others = data.drop(columns=['id', 'chimsoo', 'rain'])
X_rain = data[['rain']]
for column in X_others.columns:
    X_rain[column + '_interaction'] = X_others[column] * data['rain']
X_train, X_test, y_train, y_test = train_test_split(X_rain, y, test_size=0.2, random_state=41)

# 선형 회귀 모델 생성 및 학습
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)# =정확도

print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
print("Coefficients (Slope):", model.coef_)
print("Intercept:", model.intercept_)

print("Please upload the CSV file containing actual test data (with 'chimsoo')")
uploaded_test = files.upload()
test_file_name = list(uploaded_test.keys())[0]
test_data = pd.read_csv(test_file_name)
actual_chimsoo = test_data['chimsoo']

#rain =  x값
test_data_others = test_data.drop(columns=['id', 'chimsoo', 'rain'])
test_data_rain = test_data[['rain']]

# 나머지 변수들로 깅울기와 y절편 생성
for column in test_data_others.columns:
    test_data_rain[column + '_interaction'] = test_data_others[column] * test_data['rain']
predicted_chimsoo = model.predict(test_data_rain)
print("Example for 20 ")
results = pd.DataFrame({
    'Actual Chimsoo': actual_chimsoo,
    'Predicted Chimsoo': predicted_chimsoo
})

print(results)
mse_test = mean_squared_error(actual_chimsoo, predicted_chimsoo)
r2_test = r2_score(actual_chimsoo, predicted_chimsoo)
accuracy = r2_test * 100
print("Mean Squared Error on Test Data:", mse_test)
print("R^2 Score on Test Data:", r2_test)
print("Accuracy on Test Data: {:.2f}%".format(accuracy))
```
